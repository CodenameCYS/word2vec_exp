```python
import pickle
from matplotlib import pyplot as plt
```

```python
color = list("红赤黄绿青蓝紫黑白灰")
place = "大理 天竺 梁山 景阳冈 洛阳 东京 江南 沧州 凉州 邯郸".split()
jobs = "知府 教头 县令 账房 丞相 太尉 皇帝 国师 员外 知州".split()
```

```python
cbow_result = pickle.load(open("data/cbow_result.pkl", "rb"))
c_points = [cbow_result[w] for w in color]
p_points = [cbow_result[w] for w in place]
j_points = [cbow_result[w] for w in jobs]
plt.figure(figsize=(15,7))
plt.scatter([x[0] for x in c_points], [x[1] for x in c_points], marker=".", color="red")
plt.scatter([x[0] for x in p_points], [x[1] for x in p_points], marker="x", color="green")
plt.scatter([x[0] for x in j_points], [x[1] for x in j_points], marker="^", color="black")
plt.show()
```

```python
import os
import tensorflow as tf

os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

```python
cbow_embedding = tf.keras.models.load_model("model/cbow_embedding.h5")
skip_gram_embedding = tf.keras.models.load_model("model/skip_gram_embedding.h5")
trival_embedding = tf.keras.models.load_model("model/trival_embedding.h5")
```

```python
vocabfile = "data/vocab.txt"
w2id = {line.strip():idx for idx, line in enumerate(open(vocabfile))}
id2w = {idx:line.strip() for idx, line in enumerate(open(vocabfile))}
```

```python
import numpy
```

```python
w1 = "林冲"
w2 = "豹子头"
inputs = numpy.array([w2id[w1], w2id[w2]])
print(cbow_embedding.predict(inputs))
print(skip_gram_embedding.predict(inputs))
print(trival_embedding.predict(inputs))
```
